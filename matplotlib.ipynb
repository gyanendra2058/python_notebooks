```python
# Load housing prices data
import matplotlib.pyplot as plt
import pandas as pd
housing_df = pd.read_csv('HousePrices.csv')
city_avg = housing_df.groupby('city')['price'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(25,15))
plt.pie(city_avg, labels=city_avg.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Cities by Average House Price')
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Disable LaTeX text rendering
plt.rcParams['text.usetex'] = False
# Fixing random state for reproducibility
np.random.seed(19680801)

fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```
