```python
a = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# To append at the start (prepend), use insert(0, item)
a.insert(0, 'PreMonth')
print(a)

# Alternative: a = ['PreMonth'] + a
# But this creates a new list, while insert modifies in place
```

```python
evenSum = 0;
oddSum= 0;
for i in range(1,101):
    if(i%2 == 0):
        evenSum += i
    else:
        oddSum += i
print("Even sum:", evenSum)
print("Odd Sum:", oddSum)
```
