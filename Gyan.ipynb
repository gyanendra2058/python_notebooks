```python
def draw_stars(count):
    result =""
    for i in range(0, count):
        result+=" *"
    return result;

def draw_pattern(): 
    for i in range(1,10):
        if(i<=5): 
            print(draw_stars(i))
        else: print(draw_stars(10-i))
        
draw_pattern()
```

```python
def myFun():
    my_list = [1, 2, 3, 'Jessa', 4, 5, 'Kelly', 'Jhon', 6,9]
    new_list = []
    for el in my_list:
        if(type(el) == int):
            new_list.append(el)
    
    return new_list

print(myFun())
```
